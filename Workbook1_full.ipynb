### Типы данных
Все типы данных в Python относятся к одной из 2-х категорий: 
изменяемые (mutable) и неизменяемые (immutable).

*    числовые данные (int, float),

*	списки (list),

## If

**Условный оператор**

В коде часто приходится проверять выполнимость или невыполнимость каких-то условий. Синтаксис следующий

```
if <условие1 (булевское выражение)> :
    <код, который выполнится, если условие верно>
elif <условие2 (булевское выражение)>:
    <код, который выполнится, если условие1 было неверно, а условие2 верно>
else:
    <код, который выполнится, если условие1 и условие2 были неверны>
```

Обратите внимание, что код, который должен выполняться внутри каждого условия, записывается с отступом в 4 пробела от уровня if, elif и else: в питоне области видимости переменных обозначаются отступами.

*То есть, отступы позволяют понять, где начинается код, который должен выполняться при выполнении условия в if, и где заканчивается*

Пример: Вывести на экран является ли переменная положительной, отрицательной или равна нулю. 

````python
x = 125
if x<0:
    print("x отрицательный")
elif x==0:
    print("x равен нулю")
else:
    print("x положительный")
````

## Циклы - for и while 

Цикл while проверяет истинность условия, и если условие истинно, то выполняет инструкции цикла.<br> 
Синтаксис:<br>

```
while условие:
   инструкции
```
Инструкции, которые относятся к циклу while, располагаются на последующих строках и должны иметь отступ от начала ключевого слова while.

Задание. Вывести все числа от 1 до 10 с шагом 3.

Цикл for проходит по набору значений, помещает каждое значение в переменную с которой в цикле можно производить различные действия. <br>
Синтаксис:
```
for переменная in набор_значений:
    инструкции
```

После ключевого слова for идет название переменной, в которую будут помещаться значения. <br>
После оператора in указывается набор значений и двоеточие. Инструкции, которые относятся к циклу, располагаются на последующих строках и должны иметь отступ.

4.3.5 Создайте простейший калькулятор, включающий основные действия для двух переменных '+','-','∙','÷', 
а также вычисление следующих функций: $e^{(x+y)},sin⁡{(x+y)},cos⁡{(x+y)},x^y$.

````python
import math as m
print("Ноль в качестве знака операции"
      "\nзавершит работу программы")
while True:
    s = input("Знак (+,-,*,/, sin, cos, e^(x+y), x^y): ")
    if s == '0':
        break
    if s in ('+', '-', '*', '/', 'sin','cos', 'e^(x+y)', 'x^y'):
        x = float(input("x="))
        y = float(input("y="))
        if s == 'sin':
            print (m.sin(x+y))
        elif s == 'cos':
            print (m.cos(x+y))
        elif s == 'e^(x+y)':
            print (m.exp(x+y))
        elif s == 'x^y':
            print(m.pow(x,y))
        elif s == '+':
            print("%.2f" % (x+y))
        elif s == '-':
            print("%.2f" % (x-y))
        elif s == '*':
            print("%.2f" % (x*y))
        elif s == '/':
            if y != 0:
                print("%.2f" % (x/y))
            else:
                print("Деление на ноль!")
    else:
        print("Неверный знак операции!")
````
